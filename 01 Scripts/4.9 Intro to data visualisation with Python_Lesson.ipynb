# 07 Creating a representative subset through sampling

```python
# Using a non-representative subset to understand that this does not work
bad_sub = ords_prods_merge[:2000000]
```

```python
# Creating a line chart showing prices of items bought across the week with the non-representative subset
sns.lineplot(data=bad_sub, x='order_day_of_the_week', y='prices')
```

Comparing this to the line chart of the full dataset immediately shows that using a non-representative subset does not make any sense.

```python
# Creating samples according to a 70/30 split
np.random.seed(4)
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
# Using the created list (dev) to create the two samples according to the 70/30 split
sub_70 = ords_prods_merge[dev]
sub_30 = ords_prods_merge[~dev]
```

```python
# Checking length of 70% subset
len(sub_70)
```

```python
# Checking length of 30% subset
len(sub_30)
```

```python
# Checking whether length of full dataset is identical with sum of length of subsets
len(ords_prods_merge) == len(sub_70) + len(sub_30)
```

```python
# Further subsetting to only include columns on prices and order_day_of_the_week
sub_sub_30 = sub_30[['order_day_of_the_week', 'prices']]
```

```python
# Creating line chart showing prices of items bought across the week with 30% representative subset
line2 = sns.lineplot(data=sub_sub_30, x='order_day_of_the_week', y='prices')
```

The line chart from the sub_sub_30 dataset looks very similar to the line chart obtained from the full dataset. This confirms that the sub_sub_30 subset is a sample, i.e. a representative subset of the full dataset.

```python
# Exporting linecharts of full dataset and sample
line1.figure.savefig(os.path.join(path, '04 Analysis', '02 Visualisations', 'line_orderdow_prices_full.png'))
line2.figure.savefig(os.path.join(path, '04 Analysis', '02 Visualisations', 'line_orderdow_prices_sample.png'))
```

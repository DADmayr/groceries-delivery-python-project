## Step 4: Creating a loyalty flag
If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”
If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”
If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”

```python
# Creating new column with maximum number of orders per customer
ords_prods_merge['maximum_orders'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking dimensions of dataframe with new column
ords_prods_merge.shape
```

```python
# Checking head of dataframe with new column
ords_prods_merge.head(20)
```

```python
# Checking selected columns only
ords_prods_merge[['user_id', 'order_number', 'maximum_orders']].head(20)
```

```python
# Adding loyalty flag in new column with conditions I-III
ords_prods_merge.loc[ords_prods_merge['maximum_orders'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['maximum_orders'] > 10) & (ords_prods_merge['maximum_orders'] <= 40), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['maximum_orders'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking flag assingment and counting labels
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

Approx. 49% of the customers are "regular customers". Loyal customers make up approx. 32% of the customer base.

## Step 5: Comparing prices of products purchased by new, regular and loyal customers

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['min', 'mean', 'median', 'max']})
```

Minimum, maximum and median are identical for the three groups of clients. However, the average price is highest vor new customers and lowest for loyal customers. This could point towards loyal customers knowing better which products to purchase cheaply at Instacart and from which products to stay away because they're expensive. It could also point towards new customers simply wanting to get specific products delivered with the price of those products not mattering to them so much. 
At this point, this is only guessing. A deeper look into the products purchased would be necessary to make qualified statements about reasons for the identified pattern.

## Step 6: Creating a spending flag
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Creating new column with average price of products purchased per user
ords_prods_merge['average_price_for_user'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking head of dataframe with new column
ords_prods_merge.head()
```

```python
# Checking head for selected columns only
ords_prods_merge[['user_id', 'prices', 'average_price_for_user']].head(20)
```

```python
# Creating a spending flag with conditions I and II
ords_prods_merge.loc[ords_prods_merge['average_price_for_user'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['average_price_for_user'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Checking assingment of flags and counting labels
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
# Counting users
ords_prods_merge.groupby('spending_flag')['user_id'].nunique()
```

Only 2,6% of the customers are high spenders.

## Step 7: Creating an order frequency flag to categorise frequent vs. non-frequent users
Point of comparison is the median in the “days_since_prior_order” column.
If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Creating new column with median "days since prior order" per user
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Checking dataframe with new column
ords_prods_merge.head()
```

```python
# Checking specific columns only
ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_since_prior_order']].head(20)
```

```python
# Creating order_frequency_flag
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] > 10) & (ords_prods_merge['median_days_since_prior_order'] <= 20), 'order_frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Checking assignment of flags and counting values
ords_prods_merge['order_frequency_flag'].value_counts(dropna=False)
```

```python
# Counting users
ords_prods_merge.groupby('order_frequency_flag')['user_id'].nunique()
```

Approx. 50,2% of the customers are frequent customers, approx. 20,9% are non-frequent customers, and approx. 28,8% are regular customers.
